## THE OO METHOD

```python
over_50=heart_disease[heart_disease["age"]>50]
over_50.head()
```

```python
over_50.plot(kind="scatter",
             x="age",
             y="chol",
            c='target')
```

```python
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
#ax.set_xlim([45,100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);
ax.set(title="heart disease and cholestrol level",
      xlabel="age",
      ylabel="cholestrol");
ax.legend(*scatter.legend_elements(),title="Title");
ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
over_50.target.values
```

```python
#subplot of age ,thalach
fig,(ax0,ax1)=plt.subplots(nrows=2,
                           ncols=1,
                           figsize=(10,6),
                           sharex=True)
scatter=ax0.scatter(x=over_50["age"],
                   y=over_50["chol"],
                  c=over_50["target"]);
ax0.set(title="AGE nd cholestrol",
       ylabel="cols");
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(y=over_50["chol"].mean(),linestyle="--");

scatter=ax1.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                  c=over_50["target"]);
ax1.set(title="AGE nd HEART RATE",
       xlabel="age",
       ylabel="thalach");
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="-");
fig.suptitle("Heart Disease Analysis",fontsize=16,fontweight="bold");
```
